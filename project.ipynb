## Analyzing the relation between the variable

```python
corr = df.corr()

top_corr = corr["Class"].abs().sort_values(ascending=False)


top_features = top_corr[1:11].index  


filtered_corr = df[top_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(filtered_corr, annot=True, cmap="coolwarm", center=0)
plt.title("Heatmap of Top Correlated Features with Class")
plt.show()
```

### Data is too much it will take too much time to visualize so we will simply find the top most important features using the random forest

```python
from sklearn.ensemble import RandomForestClassifier


X = df.drop("Class", axis=1)
y = df["Class"]

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

importances = model.feature_importances_
feat_importance = pd.Series(importances, index=X.columns)
feat_importance.nlargest(15).plot(kind='barh')
plt.title("Top 15 Important Features")
plt.show()
```

```python
#storing it in the x
top_features = feat_importance.nlargest(10).index
x = X[top_features]
```

## Removing the imbalance in the data so we can have a better look and result

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
smote = SMOTE(random_state= 42)
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

```python
y_train_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,f1_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    "LogisticRegression": LogisticRegression(),
    "KNN": KNeighborsClassifier(  n_neighbors=5,      
    weights='uniform', 
    metric='minkowski' ),
    "Naive": GaussianNB(),
    "Decision": DecisionTreeClassifier(    max_depth=5,              # Controls overfitting
    min_samples_split=10,    
    min_samples_leaf=5,       
    random_state=42),
    "svm": SVC(    kernel='rbf',         
    C=1.0,                
    gamma='scale',        
    probability=True  ),
    "tree": RandomForestClassifier()
}
```

```python
result = []
```

```python
x_train_sm_sample = x_train_sm.sample(n=5000, random_state=42)
y_train_sm_sample = y_train_sm.loc[x_train_sm_sample.index]
```

```python
for name, model in models.items():
	model.fit(x_train_sm_sample, y_train_sm_sample)
	y_pred = model.predict(x_test)
	acc = accuracy_score(y_test, y_pred)
	f1 = f1_score(y_test, y_pred)
	result.append({
		'model': name,
		"Accuracy": round(acc, 4),
		'f1 score': round(f1,4)
	})
  
```

```python
result
```

```python
y_train_sm_sample.value_counts()
```

```python
x_test_sample = x_test.sample(n=5000, random_state=42)
y_test_sample = y_test.loc[x_test_sample.index]
```

```python
# Step 1: Train your final model on the full balanced training set
models['svm'].fit(x_train_sm_sample, y_train_sm_sample)

# Step 2: Predict on real unseen test set
y_test_pred = models['svm'].predict(x_test_sample)

# Step 3: See how it performs
from sklearn.metrics import classification_report
print(classification_report(y_test_sample, y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_sample, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_predict, StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

y_pred = cross_val_predict(models['svm'], x_train_sm_sample, y_train_sm_sample, cv=cv)
```

```python
print(classification_report(y_train_sm_sample, y_pred))
```
